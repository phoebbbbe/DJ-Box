```python
import numpy as np
import pandas as pd
import json
import librosa
import math
import os
import soundfile as sf
```

### 將 wav 檔切割成好幾個 60sec

```python
# 資料集路徑、csv路徑
label = "occasion"
dataset_path = f'/Users/phoebe/DJ-Box_model/dataset/music_dataset/{label}_set'
train_dataset_path = f'/Users/phoebe/DJ-Box_model/dataset/music_train_dataset/{label}_train_set'

# 先將wav分成好幾個60sec的wav
DUR = 60
SEG = 10
SR = 22050
```

```python
# 遍歷整個資料集
for i, (dir_path, dir_names, file_names) in enumerate(os.walk(dataset_path)):
    
    if dir_path is not dataset_path:
        
        dir_path_folders = dir_path.split("/")
        semantic_label = dir_path_folders[-1]
        for file in sorted(file_names):
            print(f"{file}")
            
            if file == '.DS_Store':
                continue
            else:
                
                # 載入聲音資料
                file_path = os.path.join(dir_path, file)
                signal, sample_rate = librosa.load(file_path, sr=SR)
                
                # 去除聲音兩端的無聲部分
                signal, _ = librosa.effects.trim(signal)
                
                # 將聲音資料切割成相同長度
                signal_duration = librosa.get_duration(y=signal, sr=sample_rate)
                
                n_parts = int(signal_duration / DUR)
                part_len = SR * DUR
                
                for i in range(n_parts):
                    start = i * part_len
                    finish = start + part_len
                    part = signal[start:finish]
                    sf.write(f'{train_dataset_path}/{semantic_label}/{file[:-4]}_{i}.wav', part, SR)
```
